# Neural Network

<span>

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import accuracy_score
from DataTreatmentPipeline import pipeline_processing
from sklearn.preprocessing import StandardScaler
```

```python
data_train = pd.read_csv("Titanic/Titanic_train_processed.csv")
y_train = data_train["Survived"]
X_train = data_train.drop(columns=["Survived"])

# Validation test before test
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# We need to standardize for a neural network
scaler = StandardScaler()
X_tr_scaled = scaler.fit_transform(X_tr)
X_val_scaled = scaler.transform(X_val)

# Model
model = Sequential([
    Dense(32, activation='relu', input_shape=(X_tr_scaled.shape[1],)),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X_tr_scaled, y_tr,
    validation_data=(X_val_scaled, y_val),
    epochs=50,
    batch_size=32,
    verbose=1
)
```

## Test dataset

<br>

<span>

```python
# Loading the dataset
data = pd.read_csv("Titanic/test.csv")

# Processing the dataset
pipeline_processing(data)

# Loading the data we want to predict
data_test = pd.read_csv("Titanic/Titanic_test_processed.csv")
data_test_scaled = scaler.transform(data_test)

# Prediction
y_test_pred = model.predict(data_test_scaled)
y_test_pred = (y_test_pred > 0.5).astype(int).ravel()
```

```python
# Submission for Kaggle
submission = pd.DataFrame({
    "PassengerId": data_test["PassengerId"],
    "Survived": y_test_pred
})

submission.to_csv("submission_neural_network.csv", index=False)
```
